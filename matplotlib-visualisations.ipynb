# Subplot()

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# set up a subplot grid that has height 2 and has width 1
# and set the first such an active subplot
plt.subplot(2,1,1)
# make the first plot
plt.plot(x,y_sin)
plt.title('SINE')
# MAKE THE second subplot as active, make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('COSE')
# SHOW the figure
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# set up a subplot grid that has height 2 and has width 1
# and set the first such an active subplot
plt.subplot(2,1,1)
# make the first plot
plt.plot(x,y_sin,'b--')
plt.title('SINE')
# MAKE THE second subplot as active, make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'y--')
plt.title('COSE')
# SHOW the figure
plt.show()
```

```python
# BOXPLOT using MATPLOTLIB
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True,patch_artist=True)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True,patch_artist=False)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=False,patch_artist=True)
```

```python
labels='Python','Java','C++','R'
sizes=[1222,445,66,88]
colors=['gold','orange','green','blue']
explode=(0.4,0.1,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
